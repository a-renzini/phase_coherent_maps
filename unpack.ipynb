```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
!ls -ltr
```

# constant phase across the sky; nside_in = 8, nside_out = 4

```python
filein = np.load('const_ph_tests/hp_hc_in_ns8_cp.npz') 
file   = np.load('const_ph_tests/map_PC_cp_ns_4.npz')
```

```python
hp_in = filein['hp_in']
hc_in = filein['hc_in']

hp_out = file['h_plus']
hc_out = file['h_cross']
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(figsize = (18,10), ncols=4)

plt.axes(ax1)
hp.mollview(hp_in.real, hold = True)
plt.axes(ax2)
hp.mollview(hp_in.imag, hold = True)
plt.axes(ax3)
hp.mollview(hc_in.real, hold = True)
plt.axes(ax4)
hp.mollview(hc_in.imag, hold = True)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(figsize = (18,10), ncols=4)

plt.axes(ax1)
hp.mollview(hp_out.real, hold = True)
plt.axes(ax2)
hp.mollview(hp_out.imag, hold = True)
plt.axes(ax3)
hp.mollview(hc_out.real, hold = True)
plt.axes(ax4)
hp.mollview(hc_out.imag, hold = True)
```

```python
fig, (ax1, ax2) = plt.subplots(figsize = (18,10), ncols=2)

plt.axes(ax1)
hp.mollview(hp_in*np.conj(hp_in) +hc_in*np.conj(hc_in), hold = True)
plt.axes(ax2)
hp.mollview(hp_out*np.conj(hp_out) +hc_out*np.conj(hc_out), hold = True)
```
