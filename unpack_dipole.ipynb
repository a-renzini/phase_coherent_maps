# Intensity 2pole mapping

```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
!ls -ltr
```

# nside_in = 4, nside_out = 4

```python
filein = np.load('hp_hc_in_ns16_2pole.npz') 
file   = np.load('map_2poleHL_ns_4.npz')

hp_in = filein['hp_in']
hc_in = filein['hc_in']

hp_out1 = file['h_plus']
hc_out1 = file['h_cross']

map_in  = hp_in*np.conj(hp_in) +hc_in*np.conj(hc_in)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(figsize = (18,10), ncols=4)

plt.axes(ax1)
hp.mollview(hp_in.real, hold = True, title ="hp_in real")
plt.axes(ax2)
hp.mollview(hp_in.imag, hold = True, title ="hp_in imag")
plt.axes(ax3)
hp.mollview(hc_in.real, hold = True, title ="hc_in real")
plt.axes(ax4)
hp.mollview(hc_in.imag, hold = True, title ="hc_in imag")


fig, (ax1, ax2, ax3, ax4) = plt.subplots(figsize = (18,10), ncols=4)

plt.axes(ax1)
hp.mollview(hp_out1.real, hold = True, title ="hp_in real")
plt.axes(ax2)
hp.mollview(hp_out1.imag, hold = True, title ="hp_in imag")
plt.axes(ax3)
hp.mollview(hc_out1.real, hold = True, title ="hc_in real")
plt.axes(ax4)
hp.mollview(hc_out1.imag, hold = True, title ="hc_in imag")


##
map_out = hp_out1*np.conj(hp_out1) + hc_out1*np.conj(hc_out1)


fig, (ax1, ax2) = plt.subplots(figsize = (18,10), ncols=2)

plt.axes(ax1)
hp.mollview(map_in, hold = True, title = 'input')
plt.axes(ax2)
hp.mollview(map_out, hold = True, title = 'NP')

print(np.mean(hp_in*np.conj(hp_in) +hc_in*np.conj(hc_in)), np.mean(hp_out1*np.conj(hp_out1) +hc_out1*np.conj(hc_out1)))
```
