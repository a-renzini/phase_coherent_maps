# Intensity point source mapping

```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
!ls -ltr
```

# nside_in = 4, nside_out = 4

```python
filein = np.load('GB_point_source/hp_hc_in_ns16_GB_np.npz') 
file   = np.load('GB_point_source/Imap_GB_HL_ns_4.npz')

hp_in = filein['hp_in']
hc_in = filein['hc_in']

map_out = file['map']
map_in  = hp_in*np.conj(hp_in) +hc_in*np.conj(hc_in)

fig, (ax1, ax2) = plt.subplots(figsize = (18,10), ncols=2)

plt.axes(ax1)
hp.mollview(hp.ud_grade(map_in, nside_out = 4), hold = True, title = 'Intensity in')
plt.axes(ax2)
hp.mollview(map_out, hold = True, title = 'Intensity out')
```

# nside_in = 16, nside_out = 8

```python
file    = np.load('GB_point_source/Imap_GB_HL_ns_8.npz')
map_out = file['map']

fig, (ax1, ax2) = plt.subplots(figsize = (18,10), ncols=2)

plt.axes(ax1)
hp.mollview(map_in, hold = True, title = 'Intensity in')
plt.axes(ax2)
hp.mollview(map_out, hold = True, title = 'Intensity out')

print(np.mean(map_in), np.mean(map_out))
```

```python
hp.mollview(hp.ud_grade(map_in, nside_out = 8), hold = True, title = 'Intensity in udgraded')
```

```python
fileGBi1 = np.load('GB_point_source/hp_hc_in_ns16_GB_np_smooth.npz')
fileGBo1 = np.load('output_maps/map_GB_pc_smoothHL_ns_8.npz')

hp_i1 = hp.ud_grade(fileGBi1['hp_in'], nside_out = 16)
hc_i1 = hp.ud_grade(fileGBi1['hc_in'], nside_out = 16)

hp_o1 = fileGBo1['h_plus']
hc_o1 = fileGBo1['h_cross']
import matplotlib.pyplot as plt
from matplotlib import rc
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
import matplotlib as m
m.rcParams.update({'font.size':18})
from matplotlib import cm
gnu_cmap = cm.gnuplot
gnu_cmap.set_under("w")
mag_cmap = cm.magma
mag_cmap.set_under("w")

#from matplotlib import rc
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
#rc('text', usetex=True)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(figsize = (18,10), ncols=4)
#plt.title('Coherent point source input')
plt.axes(ax1)
hp.mollview(hp_i1.real, hold = True, min = 0., max = 0.18, title =r"$h^+_{\mathbb{R}}$", cmap = mag_cmap)
plt.axes(ax2)
hp.mollview(hp_i1.imag, hold = True, title =r"$h^+_{\mathbb{I}}$", cmap = mag_cmap)
plt.axes(ax3)
hp.mollview(hc_i1.real, hold = True, min = 0., max = 0.18, title =r"$h^\times_{\mathbb{R}}$", cmap = mag_cmap)
plt.axes(ax4)
hp.mollview(hc_i1.imag, hold = True, min = 0., title =r"$h^\times_{\mathbb{I}}$", cmap = mag_cmap)

plt.savefig('GB_point_source_coherent_in.pdf', bbox_inches='tight')

fig, (ax1, ax2, ax3, ax4) = plt.subplots(figsize = (18,10), ncols=4)

plt.axes(ax1)
hp.mollview(hp_o1.real, hold = True, min = 0., max = 0.17, title =r"$h^+_{\mathbb{R}}$", cmap = mag_cmap)
plt.axes(ax2)
hp.mollview(hp_o1.imag, hold = True, min = -0.01, max = 0.01, title =r"$h^+_{\mathbb{I}}$", cmap = mag_cmap)
plt.axes(ax3)
hp.mollview(hc_o1.real, hold = True, min = 0., max = 0.17, title =r"$h^\times_{\mathbb{R}}$", cmap = mag_cmap)
plt.axes(ax4)
hp.mollview(hc_o1.imag, hold = True, min = -0.01, max = 0.01, title =r"$h^\times_{\mathbb{I}}$", cmap = mag_cmap)

unit_text_obj = CbAx.get_children()[1]
unit_text_obj.set_fontsize(20)
unit_text_obj.set_y(-1.5)

plt.savefig('GB_point_source_coherent_out.pdf', bbox_inches='tight')

##

fig, (ax1, ax2) = plt.subplots(figsize = (18,10), ncols=2)

plt.axes(ax1)
hp.mollview(hp_i1*np.conj(hp_i1) +hc_i1*np.conj(hc_i1), hold = True, min = 0, max = 0.06, title = 'Input Intensity', cmap = mag_cmap)
plt.axes(ax2)
hp.mollview(hp_o1*np.conj(hp_o1) +hc_o1*np.conj(hc_o1), hold = True, min = 0, max = 0.05, title = 'Reconstructed Intensity', cmap = mag_cmap)

f = plt.gcf().get_children()
HpxAx = f[1]
CbAx = f[2]

#coord_text_obj = HpxAx.get_children()[0]
#coord_text_obj.set_fontsize(fontsize)

unit_text_obj = CbAx.get_children()[1]
unit_text_obj.set_fontsize(20)
unit_text_obj.set_y(-1.5)

plt.savefig('GB_point_source_coherent_I.pdf', bbox_inches='tight')

print(np.mean(hp_i1*np.conj(hp_i1) +hc_i1*np.conj(hc_i1)), np.mean(hp_o1*np.conj(hp_o1) +hc_o1*np.conj(hc_o1)))
```
